```python
# Filling missing values in the 'Year' column with the mode
df['Year'] = df['Year'].fillna(df['Year'].mode()[0])

# Filling missing values in the 'Gender' column with the mode
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])

# Filling missing values in the 'Age_group' column with the mode
df['Age_group'] = df['Age_group'].fillna(df['Age_group'].mode()[0])

# Check the remaining missing values after imputation
filling_first = df.isnull().sum()
print("Remaining Missing Values after Imputation:\n", filling_first)

# Filling missing values in the 'Total' column with 0
df['Total'].fillna(0, inplace=True)

# Calculating the total number of missing values after filling missing values
total_missing = df.isnull().sum().sum()
print("Total Number of Missing values after filling missing values:", total_missing)
```

2. For your selected two variables plot a figure to analyze Skewness and Modality of Distributions
present in dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Get the counts of each 'Type_code' in the DataFrame
genre_counts = df['Type_code'].value_counts()

# Set a threshold for less frequent 'Type_code'
threshold = 50

# Identify 'Type_code' that appear less frequently than the threshold
less_frequent_genres = genre_counts[genre_counts < threshold].index

# Aggregate less frequent 'Type_code' into an 'Other' category
df['Type_code'] = df['Type_code'].apply(lambda x: 'Other' if x in less_frequent_genres else x)

# Set up subplots
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(14, 6))

# Create a histogram (with KDE) for the 'Type_code' column with a different color
sns.histplot(df['Type_code'], kde=True, ax=axes, color="Skyblue")
axes.set_title('Distribution of Type Codes')
axes.set_xlabel('Type Code')

# Calculate skewness and determine modality for 'Type_code'
skewness = df['Type_code'].value_counts().skew()
modality = 'Unimodal' if skewness > -1 and skewness < 1 else 'Multimodal'
axes.text(0.5, -0.15, f'Skewness: {skewness:.2f}, Modality: {modality}', transform=axes.transAxes, ha='center')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Get numerical columns in the dataset
numerical_columns = df.select_dtypes(include='number').columns

# Create subplots with boxplots for each numerical column
fig, axes = plt.subplots(nrows=1, ncols=len(numerical_columns), figsize=(16, 6))

for i, column in enumerate(numerical_columns):
    sns.boxplot(x=column, data=df, color='skyblue', ax=axes[i])
    axes[i].set_title(f'Boxplot for {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Values')

# Add a main title to the figure
fig.suptitle('Boxplots for Numerical Variables', fontsize=16)

plt.tight_layout()
plt.show()
```

1. Explore correlation visualization with scatter plot between the variables. {Hint: Explore the seborn pairplot function.}

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your CSV file into a DataFrame
df = pd.read_csv('Suicides-in-India-2001-2012.csv')

# Selecting a subset of numeric columns from the DataFrame
selected_numeric_columns = df[['Year', 'Total']]  # Selecting 'Year' and 'Total' columns

# Sampling a fraction of the dataset
sampled_dataset = selected_numeric_columns.sample(frac=0.01)

# Define a custom color palette
custom_palette = sns.color_palette("Set3")

# Create a pairplot for the sampled numeric columns
sns.pairplot(sampled_dataset, diag_kws={'color': custom_palette[1]}, plot_kws={'s': 10, 'color': custom_palette[0]})

# Setting title for the pairplot
plt.suptitle('Pairplot of Selected Numeric Variables', y=1.02)

# Show the plot
plt.show()

```

```python
# Create a correlation matrix
correlation_matrix = df.corr()

# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
